```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
import pickle
```

```python
df=pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df
```

```python


encoder =LabelEncoder()

df['Gender']=encoder.fit_transform(df['Gender'])
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder_geography = OneHotEncoder()
geo_encoded=encoder_geography.fit_transform(df[['Geography']])
geo_encoded
```

```python
geo_encoded.toarray()
```

```python
geo_encoded_df=pd.DataFrame(geo_encoded.toarray(),columns=encoder_geography.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
df=pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
df.head()
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(encoder,file)

with open('one_encoder_geography.pkl','wb') as file:
    pickle.dump(encoder_geography,file)
```

```python
x =df.drop('Exited',axis=1)
y =df['Exited']
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,train_size=0.20,random_state=22)
scaler =StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
from tensorflow.keras.layers import Dense
import datetime
```

```python
model =Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
loss=tf.keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
log_dir= 'log/fit'
tensor_board=TensorBoard(log_dir=log_dir,histogram_freq=1)
callback =EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history = model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensor_board,callback]
)
```

```python
model.save('ann.h5')
```
