```python
import tensorflow as tf
import pandas as pd
import pickle
import numpy as np
from tensorflow.keras.models import load_model
```

```python
model =load_model('ann.h5')

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder=pickle.load(file)

with open('one_encoder_geography.pkl','rb') as file:
    one_encoder=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_encoded=one_encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=one_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df =pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender'] =label_encoder.transform([input_data['Gender']])
```

```python
data=pd.concat([input_df.drop(['Geography'],axis=1),geo_encoded_df],axis=1)
```

```python
data=scaler.transform(data)
```

```python
data
```

```python
predection =model.predict(data)
predection
```

```python
predection_prob=predection[0][0]
```

```python
if(predection_prob>0.5):

    print("this is likely to churn")
else:
    print("this is not likely to churn")
```
